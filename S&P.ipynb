```python
import pandas as pd
import matplotlib.pyplot as plt
import datetime as dt

spx_df = pd.read_csv('/content/SPX.csv', parse_dates=['Date'])

#Taking a first look
spx_df.head()
spx_df.tail()
```

```python
spx_df.describe()
```

```python
#Basic Stuff
spx_df.dropna()
spx_df.duplicated()
spx_df.drop_duplicates(inplace=True)
```

How about a nice visual representation of the S&P500







```python
plt.plot(spx_df['Date'], spx_df['Close'], label='Close')
plt.plot(spx_df['Date'], spx_df['Open'], label='Open')
plt.title('S&P 500 from 1928-2020')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# Setting xticks to every 4 years
spx_dates = pd.date_range(start= spx_df.iloc[0]['Date'], end= spx_df.iloc[len(spx_df)-1]['Date'], freq='4YS')
plt.xticks(spx_dates, spx_dates.strftime('%Y'),rotation=45)
plt.show()

spx_df_2000 = spx_df[spx_df['Date'] >= '1999-12-31']
plt.plot(spx_df_2000['Date'], spx_df_2000['Close'], label='Close')
plt.plot(spx_df_2000['Date'], spx_df_2000['Open'], label='Open')
plt.title('S&P 500 from 2000-2020')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# Setting xticks to every year
spx_dates = pd.date_range(start= spx_df_2000.iloc[0]['Date'], end= spx_df_2000.iloc[len(spx_df_2000)-1]['Date'], freq='YS')
plt.xticks(spx_dates, spx_dates.strftime('%Y'),rotation=45)
plt.show()



spx_df_2020 = spx_df[spx_df['Date'] >= '2019-12-31']
plt.plot(spx_df_2020['Date'], spx_df_2020['Close'], label='Close')
plt.plot(spx_df_2020['Date'], spx_df_2020['Open'], label='Open')
plt.title('S&P 500 in 2020')
plt.xlabel('Date')
plt.ylabel('Price')
plt.legend()

# Setting xticks to every Month
spx_dates = pd.date_range(start= spx_df_2020.iloc[0]['Date'], end= spx_df_2020.iloc[len(spx_df_2020)-1]['Date'], freq='MS')
plt.xticks(spx_dates, spx_dates.strftime('%m'),rotation=45)
plt.show()
```
